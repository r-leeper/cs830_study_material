# 2.27: Lab: Logistic Regression

1. Load the iris data https://scikit-learn.org/stable/datasets/toy_dataset.html#iris-plants-dataset

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris["data"]
y = iris["target"]
```

2. Build a Logistic Regression multiclass classifier to predict all three categories of flower.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {"C": [0.01, 0.1, 1, 5, 10, 15, 50, 100]}
grid_search = GridSearchCV(LogisticRegression(random_state=13, max_iter=1000), param_grid, cv=5)
grid_search.fit(X_train, y_train)

softmax_logreg = grid_search.best_estimator_
```

```python
predicts = softmax_logreg.predict(X_test)
```

3. Examine the performance using a confusion matrix.

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
ConfusionMatrixDisplay.from_predictions(y_test, predicts, display_labels=iris.target_names)
plt.show()
```
