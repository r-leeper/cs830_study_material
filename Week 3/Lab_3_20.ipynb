# Decision Tree Classifiction

1. Build a decision tree classifier for the Wine dataset

```python
# Imports
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Load wine variable
wine = datasets.load_wine()

X = wine["data"]
y = wine["target"]
```

```python
# Train test split
X_wine_train, X_wine_test, y_wine_train, y_wine_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=13,
    stratify=y
)
```

```python
# Instantiate classifier
dt_clf = DecisionTreeClassifier(max_depth=4)

# Train classifier
dt_clf.fit(X_wine_train, y_wine_train)

# Predictions
y_pred = dt_clf.predict(X_wine_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

fig = plt.figure(figsize=(18, 9))
plot_tree(dt_clf, filled=True, fontsize=12)
plt.show()
```

Lets see results

```python
print(classification_report(
    y_wine_test,
    y_pred,
    target_names=wine["target_names"]
))
```

```python
# Other import
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(
    y_wine_test,
    y_pred,
    display_labels=wine["target_names"]
)

plt.title("Wine Dataset SVM Test Set Confusion Matrix")
plt.show()
```
