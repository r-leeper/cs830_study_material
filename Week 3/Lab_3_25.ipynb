I'm not going to use the Boston housing dataset because it's actually been removed from modern release of scikit learn. Instead I'll use california housing dataset.

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html

```python
# Import necessary libraries
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load the data
housing = fetch_california_housing()
print(housing.DESCR)
```

```python
# Assign X and y
X, y = housing.data, housing.target
```

```python
# train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=13
)
```

```python
# Build and evaluate a simple regression decision tree
tree_reg = DecisionTreeRegressor(
    max_depth=8,
    min_samples_leaf=5,
    min_samples_split=10,
    random_state=13
)

tree_reg.fit(X_train, y_train)

y_pred = tree_reg.predict(X_test)

tree_mse = mean_squared_error(y_test, y_pred)
tree_rmse = tree_mse ** 0.5
tree_mae = mean_absolute_error(y_test, y_pred)
tree_r2 = r2_score(y_test, y_pred)

print("Decision Tree Regression Results")
print(f"MSE: {tree_mse:.4f}")
print(f"RMSE: {tree_rmse:.4f} (in units of $100k)")
print(f"MAE: {tree_mae:.4f}")
print(f"R²: {tree_r2:.4f}")
```
