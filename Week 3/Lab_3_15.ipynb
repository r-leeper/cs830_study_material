Try building a SVM regression model for the Boston housing dataset - https://scikit-learn.org/stable/datasets/toy_dataset.html:

I'm not going to use the Boston housing dataset because it's actually been removed from modern release of scikit learn. Instead I'll use california housing dataset.

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html

```python
# Imports

import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load the data
housing = fetch_california_housing()
print(housing.DESCR)
```

```python
# Assign X and y
X, y = housing.data, housing.target

# Check x
X
```

```python
# train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=13
)
```

```python
# Apply scaling - fit off train ONLY
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svr_model = SVR(kernel="rbf", C=100, epsilon=0.2, gamma="scale")
svr_model.fit(X_train_scaled, y_train)
```

```python
y_pred = svr_model.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae  = mean_absolute_error(y_test, y_pred)
r2   = r2_score(y_test, y_pred)

print(f"RMSE : {rmse:.4f}  (in units of $100k)")
print(f"MAE : {mae:.4f}")
print(f"R-squared : {r2:.4f}")
```
