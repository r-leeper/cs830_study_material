## Building an Ensemble Classifier

1. Build an ensemble classifier similar to the one above and apply it to the iris dataset. Remember this is a multi-class problem so you will need to adapt some of the models to accommodate this.

2. Examine the performance of the individual models and the ensemble model over the training and test data.

```python
# Standard imports
import numpy as np
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split
```

```python
# Load the iris dataset
iris = datasets.load_iris()
X = iris["data"]
y = iris["target"]
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13, stratify=y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
```

```python
knn_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier())
])

tree_pipe = Pipeline([
    ("tree", DecisionTreeClassifier(max_depth=6, random_state=13))
])

gnb_pipe = Pipeline([
    ("gnb", GaussianNB())
])

log_reg_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("log_reg", LogisticRegression(max_iter=1000, random_state=13))
])

hard_voting_clf = VotingClassifier(
    estimators=[
        ("knn", knn_pipe),
        ("decision_tree", tree_pipe),
        ("gaussian_nb", gnb_pipe),
        ("logistic_regression", log_reg_pipe)
    ],
    voting="hard"
)
```

```python
models = {
    "KNN": knn_pipe,
    "Decision Tree": tree_pipe,
    "Gaussian NB": gnb_pipe,
    "Logistic Regression": log_reg_pipe,
    "Hard Voting Ensemble": hard_voting_clf
}

for name, model in models.items():
    model.fit(X_train, y_train)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    print(f"{name}")
    print(f"  Training accuracy: {accuracy_score(y_train, train_pred):.3f}")
    print(f"  Test accuracy:     {accuracy_score(y_test, test_pred):.3f}")
```

## Using Soft Voting

```python
knn_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier())
])

tree_pipe = Pipeline([
    ("tree", DecisionTreeClassifier(max_depth=6, random_state=13))
])

gnb_pipe = Pipeline([
    ("gnb", GaussianNB())
])

log_reg_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("log_reg", LogisticRegression(max_iter=1000, random_state=13))
])

soft_voting_ensemble = VotingClassifier(
    estimators=[
        ("knn", knn_pipe),
        ("decision_tree", tree_pipe),
        ("gaussian_nb", gnb_pipe),
        ("logistic_regression", log_reg_pipe)
    ],
    voting="soft"
)
```

```python
models = {
    "KNN": knn_pipe,
    "Decision Tree": tree_pipe,
    "Gaussian NB": gnb_pipe,
    "Logistic Regression": log_reg_pipe,
    "Soft Voting Ensemble": soft_voting_ensemble
}

for name, model in models.items():
    model.fit(X_train, y_train)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    print(f"{name}")
    print(f"  Training accuracy: {accuracy_score(y_train, train_pred):.3f}")
    print(f"  Test accuracy:     {accuracy_score(y_test, test_pred):.3f}")
```
